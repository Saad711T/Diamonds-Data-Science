### Import Libraries :

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os
```

### Reading data :

```python
#read the diamonds CSV file
df = pd.read_csv('diamonds.csv')
```

```python
#display the first 10 rows of the dataframe
print(df.head(10))
```

```python
#display the last 10 rows of the dataframe
print(df.tail(10))
```

### Calculate the total of the diamonds dataset :

```python
sum = df.price.sum()
```

```python
mean = df.price.mean()
```

```python
print("The Sum of prices is :", sum)
print("The Mean of prices is :", mean)
#display the number of rows and columns in the dataframe
```

```python
descripe = df.describe()
print(descripe)
```

### Visualization :

```python
x = df.carat
y = df.price
plt.scatter(x, y, color='blue')
plt.title('Scatter plot of Carat vs Price')
plt.xlabel('Carat')
plt.ylabel('Price')
plt.show()
# Create a histogram of the price column
```

```python
plt.hist(df['price'], bins=30, color='red', edgecolor='black')
plt.title('Histogram of Price')
plt.xlabel('Price')
plt.ylabel('Frequency')
plt.show()
# Create a boxplot of the price column
```

```python
sns.boxplot(x=df['price'])
plt.title('Boxplot of Price')
plt.xlabel('Price')
plt.show()
# Create a violin plot of the price column
```
